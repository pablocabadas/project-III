```python
client = MongoClient("localhost:27017")
client.list_database_names()
db = client["Ironhack"]
companies = db.get_collection("companies")
companies
```

```python
query = {"offices.0.city": "San Francisco"}
projection = {"_id": 0, "name": 1, "category_code": 1, "founded_year": 1, "total_money_raised": 1, "offices": 1}

san_francisco= list(companies.find(query, projection))

for i in san_francisco:
    i.update({'latitude': i['offices'][0]['latitude'], 'longitude': i['offices'][0]['longitude']})
    del i['offices']

#san_francisco
df_sf = pd.DataFrame(san_francisco)
#df_sf['total_money_raised'] = (df_sf['total_money_raised'].str.replace('K', *1000).str.replace('M', *1000000)
   #                            .str.replace('B', *1000000000))
df_sf.sample(3)
#df_sf_ll=df_sf.dropna(subset=['latitude', 'longitude'], thresh=1)
#df_sf_ll   
```

```python
query = {"offices.0.city": "Bangalore", "offices.0.country_code": "IND"}
projection = {"_id": 0, "name": 1, "category_code": 1, "founded_year": 1, "total_money_raised": 1, "offices": 1}

bangalore = list(companies.find(query, projection))


for i in bangalore:
    i.update({'latitude': i['offices'][0]['latitude'], 'longitude': i['offices'][0]['longitude']})
    del i['offices']
    

df_bgl = pd.DataFrame(bangalore)
#df_bgl
df_bgl_ll=df_bgl.dropna(subset=['latitude', 'longitude'], thresh=1)
#df_bgl_ll
```

```python
query = {"offices.0.city": "Stockholm", "offices.0.country_code": "SWE"}
projection = {"_id": 0, "name": 1, "category_code": 1, "founded_year": 1, "total_money_raised": 1, "offices": 1}

stockholm = list(companies.find(query, projection))


for i in stockholm:
    i.update({'latitude': i['offices'][0]['latitude'], 'longitude': i['offices'][0]['longitude']})
    del i['offices']
    

df_stk = pd.DataFrame(stockholm)
#df_stk
df_stk_ll=df_stk.dropna(subset=['latitude', 'longitude'], thresh=1)
#df_stk_ll
```

```python
sf_map = folium.Map(location=[df_sf_ll['latitude'][0], df_sf_ll['longitude'][0]], zoom_start=14)

for i, row in df_sf_ll.iterrows():
    folium.Marker(location=[row['latitude'], row['longitude']], popup=row['name']).add_to(sf_map)

#sf_map
```

```python
#San Francisco:

    #Starbucks:
    
url = "https://api.foursquare.com/v3/places/search?ll=37.77%2C-122.42&radius=11000&chains=ab4c54c0-d68a-012e-5619-003048cad9da"
header = {"Authorization": token , "accept": "application/json" }

response = requests.get(url, headers=header)
lista_star_sf=response.json()
#lista_star_sf
```

```python
latitudes = []
longitudes = []
names = []

for i in lista_star_sf['results']:
    latitude = i['geocodes']['main']['latitude']
    latitudes.append(latitude)
    longitude = i['geocodes']['main']['longitude']
    longitudes.append(longitude)
    name = i['chains'][0]['name'] + ', ' + i['location']['address']
    names.append(name)
    
#print(latitudes, longitudes, names)
df_star_sf = pd.DataFrame ({'names': names, 'latitudes': latitudes, 'longitudes': longitudes})
df_star_sf
```
